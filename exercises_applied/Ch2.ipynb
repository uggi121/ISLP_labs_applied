```python
auto = pd.read_csv('datasets/ALL CSV FILES - 2nd Edition/Auto.csv', index_col=-1, na_values=['?']).dropna()
auto.head()
```

## a)
All variables are quantitative in the sense that they have a numerical value associated with them. However, origin, year and cylinders are discrete. Origin could be the quantitative representation of a qualitative variable. We can make the definition conservative and thus include only the below variables as quantitative:
* mpg
* displacement
* horsepower
* weight
* acceleration

```python
quantitative_predictors = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
auto[quantitative_predictors].describe().round(2)
```

```python
(auto[:10] + auto[85:])[quantitative_predictors].describe().round(2)
```

```python
pd.plotting.scatter_matrix(auto[quantitative_predictors], figsize=(12,8));
```

```python
auto.boxplot('mpg', by='cylinders')
```

```python
pd.plotting.scatter_matrix(auto, figsize=(12,8));
```

## Predicting mpg
* Quantitative predictor relationships
  * Displacement - moderate/strong negative correlation
  * Horsepower - moderate/strong negative correlation
  * Weight - moderate/strong negative correlation
  * Acceleration - moderate positive correlation
  * Year - moderate positive correlation
  * Cylinders - weak negative correlation
* Qualitative predictor relationships
  * Origin - moderate positive correlation
